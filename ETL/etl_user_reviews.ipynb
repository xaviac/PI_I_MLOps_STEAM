```python
# Bibliotecas a utilizar
import pandas as pd
import gzip
import ast
```

```python
# Se guarda el path del json en una variable
ur_path = '../data/raw/user_reviews.json.gz'

# Leer el json y guardarlo en un dataframe de pandas
filas=[]
with gzip.open(ur_path, 'rt', encoding='MacRoman') as archivo:
    for line in archivo.readlines():
        filas.append(ast.literal_eval(line))

df_user_review = pd.DataFrame(filas)

df_user_review.info()
```

```python
df_user_review.head()
```

```python
# Observar la estructura del diccionario detro de la columna reviews que parece estar dentro de []
df_user_review['reviews'][0]
```

```python
# Aplicamos explode para que cada review sea una fila
df_ur_explode = df_user_review.explode('reviews')
```

```python
df_ur_explode
```

```python
# Creamos un nuevo dataframe a partir de df_ur_explode quedando las claves del diccionario como columnas y eliminando reviews
df_ur_denested = pd.concat([df_ur_explode.drop(['reviews'], axis=1), df_ur_explode['reviews'].apply(pd.Series)], axis=1)
```

```python
df_ur_denested
```

```python
df_ur_denested.isna().sum()
```

```python
df_ur_denested.info()
```

```python
# Eliminamos columnas que no vamos a utilizar
df_ur_denested.drop(['user_url', 'funny', 'posted', 'last_edited', 'helpful', 0], axis=1, inplace=True)
```

```python
df_ur_denested.isna().sum()
```

```python
# Que porcentaje representan esos 28 registros
round(df_ur_denested.isna().sum() / len(df_ur_denested) * 100, 2)
```

```python
# Al ser un porcentaje muy bajo, eliminamos los registros con valores nulos
df_ur_denested.dropna(inplace=True)
```

```python
df_ur_denested.isna().sum()
```

```python
# Nos preparamos para hacer análisis de sentimiento una vez que tenemos nuestros datos limpios
# Importamos las librerías necesarias
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer

nltk.download('vader_lexicon')
```

```python
# Crear un objeto SentimentIntensityAnalyzer
sid = SentimentIntensityAnalyzer()
df_ur_denested['sentiment_analysis'] = df_ur_denested['review'].apply(lambda x: sid.polarity_scores(str(x))['compound'])

# Mapear los valores a la escala indicada
df_ur_denested['sentiment_analysis'] = df_ur_denested['sentiment_analysis'].apply(
    lambda score: 0 if score <= -0.5 else (2 if score >= 0.5 else 1)
)

# Si la reseña está ausente, asignar el valor 1
df_ur_denested['sentiment_analysis'].fillna(1, inplace=True)

# Mostrar el DataFrame resultante
df_ur_denested
```

```python
# Usar un for para recorrer las columnas y saber que tipo de dato es cada una
for c in df_ur_denested:
    print(c, df_ur_denested[c].apply(lambda x: type(x)).value_counts())
```

```python
# Eliminar columna review porque ya no es necesaria
df_ur_denested.drop(['review'], axis=1, inplace=True)

# Convertir la columna 'item_id' a tipo 'int32'
df_ur_denested['item_id'] = df_ur_denested['item_id'].astype('int32')
```
